```python
# Get users info from DataBase.csv to users dataframe
users = pd.read_csv('D:/DataScience/Test_de_base_DataScientist/DataBase.csv', sep=';')
```

```python
# Create output.csv as output_file
with open('D:/DataScience/Test_de_base_DataScientist/output1.csv', mode='w') as output_file:
    # Define a writer for output_file which will be used to write data to output.csv
    output_writer = csv.writer(output_file, delimiter=';')
    # Input column headers as first row in output.csv
    output_writer.writerow(['mail', 'nbAction', 'Segment'])
    # For each user in DataBase calculate required stats
    for i in range(0, len(users)):
        # Get the id of current user
        IdUser = users.loc[i, 'IdUser']
        # Get mail of current user
        mail = users.loc[i, 'mail']
        # Zero initialized variables to store total points and number of actions for current user
        pts = nbAction = 0
        # Empty variable to store segment of current user
        segment = ''
        # For each log in logs check current user id in the current log
        for j in range(0, len(logs)):
            # If current log contains current user id then calculate stats
            if logs.loc[j,'idUser'] == IdUser:
                # Increment the action as this action belongs to current user
                nbAction = nbAction + 1
                # Get the action type for the current log
                typeAction = logs.loc[j,'typeAction']

                # Assign points as per TYPE_ACTION <--> POINTS table
                if typeAction >= 0 and typeAction <= 9:
                    pts = pts + 0
                if typeAction >= 10 and typeAction <= 19:
                    pts = pts + 1
                if typeAction >= 20 and typeAction <= 29:
                    pts = pts + 2
                if typeAction >= 30 and typeAction <= 39:
                    pts = pts + 3
                if typeAction >= 40 and typeAction <= 49:
                    pts = pts + 4
                if typeAction >= 50 and typeAction <= 59:
                    pts = pts + 5

        # Assign segment as per POINTS <--> SEGMENT table
        if pts >= 0 and pts <= 4:
            segment = 'Segment 1'
        if pts >= 5 and pts <= 9:
            segment = 'Segment 2'

        if pts >= 10 and pts <= 14:
            segment = 'Segment 3'
        if pts >= 15 and pts <= 19:
            segment = 'Segment 4'
        if pts >= 20:
            segment = 'Segment 5'

        # Write results (mail, nbAction, segment) to output.csv
        output_writer.writerow([mail, nbAction, segment])

# Close the output.csv after completion of the process
output_file.close()

print('All done check output1.csv for results...')
```
