#### Dataset is collection of data which your programe can be utilise to learn, to predict new outcomes based on what it learnt(requirements)

```python
from sklearn import datasets
from sklearn import svm
#this perticular calss of sklearn has a function called trin_test_split
#which literally splits dataset into training an dtesting part
from sklearn.model_selection import train_test_split as tts

wine = datasets.load_wine()

features = wine.data
labels = wine.target
print ('Number of entreis :', len(features))
for featurename in wine.feature_names:
    print (featurename[:10],"    \t",)
    print ("Class")
for feature,label in zip(features, labels):
    for f in feature:
        print (f, "\t\t",)
        print ("label")
```

#### Predecting what calss of wine has certain set of features (here we have 3 classes and 13 features from datasets)

#### here we use SVM, a classic algorithm in ML to classify things

### Split our data for Train and Test using automatic sklearn funtion train_test_split
#### Training features corresponds to training labels and testing features corresponds to testing lebels

```python
train_feature, test_feature, train_label, test_label = tts(features, labels, test_size = 0.2)
```

```python
print("Number of entries :", len(train_feature))
for featurename in wine.feature_names:
    print(featurename[:10], "    \t",)
    print("Class")
for feature, label in zip(train_feature, train_label):
    for f in feature:
        print(f, "  \t\t",)
        print("labels")
```

#### So above 80 random things taken for training and 20 random sets taken for testing

```python
#classifier =clf  SVC linear

clf = svm.SVC(kernel="linear")
#here we are doing algorithm part
#Training
#so here svm draws a line between training features and corresponding labels 
#(as we have 3 classes & 13 features with 36 attributes in our dataset, it will shows which class belongs to which features)
#so here we provided train features and labels and saying to model that you have to learn from it
clf.fit(train_feature, train_label)

#Predictions
Predictions = clf.predict(test_feature)
print (Predictions)

#accuracy/score to check how accurate our model is
score =0
for i in range(len(Predictions)):
    if Predictions[i] == test_label[i]:
        score += 1
print (score/len(Predictions)) #to get percentage
```

### Now trying with RandomForestClassifier

### RandomForest is cluster of Decision Trees, 
### i.e It has bunch of Decision Trees and it gives output from whatever the majority happened among other DecisionTrees

#### Ex: if there are 10 decision trees, out of 8 said that these features are class 0 wine and the rest 2 said thoses features belongs to class 2 then RandomForest takes the majority's decision

```python
from sklearn import datasets
from sklearn.ensemble import RandomForestClassifier
#this perticular calss of sklearn has a function called trin_test_split
#which literally splits dataset into training an dtesting part
from sklearn.model_selection import train_test_split as tts

wine = datasets.load_wine()

features = wine.data
labels = wine.target

train_feature, test_feature, train_label, test_label = tts(features, labels, test_size = 0.2)

#classifier =clf

#This classifier takes the majority choice out of bunch of decision trees
clf = RandomForestClassifier()
#here we are doing algorithm part
#Training
#(as we have 3 classes [0 1 2] & 13 features with 36 attributes in our dataset, it will shows which class belongs to which features)
#so here we provided train features and labels and saying to model that you have to learn from it
clf.fit(train_feature, train_label)

#Predictions
Predictions = clf.predict(test_feature)
print (Predictions)

#accuracy/score to check how accurate our model is
score =0
for i in range(len(Predictions)):
    if Predictions[i] == test_label[i]:
        score += 1
print (score/len(Predictions)) #to get percentage

        
```
